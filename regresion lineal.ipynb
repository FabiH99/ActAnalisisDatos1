```python
import matplotlib.pyplot as plt
from sklearn import linear_model
import seaborn as sns
import pandas as pd
import numpy as np
import os

```

```python
datos = pd.read_csv("C:/Users/57320.LAPTOP-STBFAD2Q/Desktop/data.csv")
```

```python
datos.head(3)
```

```python
datos.columns
datos.info
datos.describe
```

```python
datos.plot.scatter(x="metro" , y="precio")
plt.show()
```

```python
regresion = linear_model.LinearRegression()
```

```python
metros = datos ["metro"].values.reshape((-1,1))
```

```python
modelo = regresion.fit(metros,datos["precio"])
print ("Interseccion (b)" , modelo.intercept_)
print ("Pendiente (m) " , modelo.coef_)
```

```python
entrada = [[7.5] , [15.0] , [17.5] , [25.0]]
predicciones = modelo.predict (entrada)
print (predicciones)
```

```python
datos.plot.scatter (x="metro" , y= "precio" , label = "Valores" )
plt.scatter (entrada, predicciones, color = "green" )
plt.plot(entrada, predicciones, color = "black" , label = "linea de regresion")
plt.xlabel('metro')
plt.ylabel('precio')
plt.legend()
plt.show()
```
